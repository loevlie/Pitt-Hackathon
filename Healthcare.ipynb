```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline


class My_Dataset(Dataset):
    
    def __init__(self,X,Y):
        self.X = X
        self.Y = Y
        
    def __len__(self):
        return len(self.Y)
    
    def __getitem__(self,index):
        X = self.X[index]
        Y = self.Y[index].long()
        return X,Y
```

```python
x = torch.Tensor(df_train.to_numpy())
y = torch.Tensor(Target.to_numpy())
```

```python
train_dataset = My_Dataset(x,y)
```

```python
train_loader_args = dict(shuffle=True,batch_size=1024,drop_last=True) 
train_loader = DataLoader(train_dataset, **train_loader_args)
```

```python
class Model(nn.Module):

    def __init__(self,in_features=72,h1=2048,h2=2048,h3 = 1028,out_features=10):
        
        # How many layers?
        # Input layer (# of features) --> hidden layer 1 (number of neurons N) --> h2 (N) --> output (346 of classes)
        super().__init__()
        self.fc1 = nn.Linear(in_features,h1)
        self.bn1 = nn.BatchNorm1d(num_features=h1,momentum=0.01)
        self.fc2 = nn.Linear(h1,h2)
        #self.d2 =  nn.Dropout(0.25)
        self.bn2 = nn.BatchNorm1d(num_features=h2,momentum=0.01)
        self.out = nn.Linear(h2,out_features)
  
    def forward(self,x):
        x = F.relu(self.bn1(self.fc1(x)))
        x = F.relu(self.bn2(self.fc2(x)))
        x = self.out(x)
        return x
```

```python
seq_model = Model()
device = torch.device("cuda")
seq_model.to(device)
```

```python
import torch 
t = torch.tensor(10)
int(t)
```

```python

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(seq_model.parameters(),lr=1e-4)
```

```python
%%time
# EPOCHS 
seq_model.train()
epochs = 25
losses = []
loss = []
val_error = []
validation_losses = []
for i in range(epochs):
    losses.append(loss)
    #validation_losses.append(val_error)
    seq_model.train()
    print(f'Epoch {i}, loss {loss}')
    for (xi,yi) in train_loader:
        optimizer.zero_grad()
        xi = xi.to(device)
        yi = yi.to(device)
        output = seq_model(xi)
        _,Y = torch.max(yi,1)
        loss = criterion(output.float(),Y)
        #loss = criterion(torch.argmax(output,1).reshape(-1,1).float(),torch.argmax(yi,1).reshape(-1,1).float())
        loss.backward()
        optimizer.step()
```
